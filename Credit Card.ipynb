*Finding Correaltion

```python
X = train_df.drop(['default_payment_next_month'],axis=1)
y = train_df['default_payment_next_month']
```

```python
X.corrwith(train_df['default_payment_next_month']).plot.bar(
        figsize = (20, 10), title = "Correlation with Default", fontsize = 20,
        rot = 90, grid = True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state = 47, 
                                    criterion = 'entropy',n_estimators=100)
classifier.fit(X_train, y_train)
```

```python
y_pred_r = classifier.predict(X_test)
from sklearn.metrics import  accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
roc=roc_auc_score(y_test, y_pred_r)
acc = accuracy_score(y_test, y_pred_r)
prec = precision_score(y_test, y_pred_r)
rec = recall_score(y_test, y_pred_r)
f1 = f1_score(y_test, y_pred_r)
results = pd.DataFrame([['Random_forest_ent100 ', acc,prec,rec, f1,roc]],
               columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score','ROC'])
results
```

```python
predictions = classifier.predict(test_df)
```

```python
submission = pd.DataFrame({'ID':test_df['ID'],'default_payment_next_month':predictions})
submission.head(10)
```

```python
filename = 'RandomForest Classifier Pred.csv'

submission.to_csv(filename,index=False)

print('Saved file: ' + filename)
```
